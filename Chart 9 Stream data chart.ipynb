```python
# Dependencies
%matplotlib notebook
import tweepy
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib import style
import time
style.use('ggplot')

# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Twitter API Keys
from config import (consumer_key, 
                    consumer_secret, 
                    access_token, 
                    access_token_secret)

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# Target User
target_user = "BBC"

compound_list = []
tweet_times = []
tweet_time_objects = []
time_in_between = []
counter=0

public_tweets = api.search(target_user, count=100, result_type="recent")
        
for tweet in public_tweets["statuses"]:
        
    results = analyzer.polarity_scores(tweet["text"])
    compound = results["compound"]
    compound_list.append(compound)
        
    tweet_times.append(tweet["created_at"])
        
    counter += 1
    if counter % 20 == 0:
        print(tweet["created_at"])
         
sentiment = {"User": target_user,"Compound": np.mean(compound_list),"Tweet Count": len(compound_list)}
BBC_compound=np.mean(compound_list)
    
for x in range(len(tweet_times)):
    tweet_datetime = datetime.strptime(tweet_times[x], "%a %b %d %H:%M:%S %z %Y")
    tweet_time_objects.append(tweet_datetime)
    if x % 20 == 0:
        print(tweet_times[x])
        print(tweet_datetime)
            
for x in range(len(tweet_time_objects)-1):
    hrs_apart = ((tweet_time_objects[x] - tweet_time_objects[x+1]).seconds) / 3600
    time_in_between.append(hrs_apart)

plt.scatter(range(len(time_in_between)+1), compound_list, marker="o",linewidth=0.25, alpha=0.8, color="r",edgecolor='k')
plt.xlim([0, len(time_in_between)])
plt.ylabel("Sentiment")
plt.xlabel("Tweets Ago")
plt.title("Tweet Sentiment Analysis")
    
```
